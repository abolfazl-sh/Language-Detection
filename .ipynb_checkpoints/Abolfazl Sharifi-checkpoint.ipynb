**Import the libraries and frameworks are needed**

```python
import os
import tarfile
import pandas as pd
import re
import string
import nltk
from unidecode import unidecode
import torch
import torchhd #using torchhd for hyperdimentional computing
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.linear_model import RidgeClassifier
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

**Data Unextraction**

```python
def unextract():
    corps_dir = "corps"
    corps = os.listdir(corps_dir)
    for corp in corps:
        for tf in os.listdir(os.path.join(corps_dir, corp)):
            ocd = os.path.join(corps_dir, corp, tf.replace(".tar.gz", ""))
            if not os.path.isdir(ocd) and "tar.gz" in ocd:
                tf_name = os.path.join(corps_dir, corp,tf)
                print(tf_name)
                tar = tarfile.open(tf_name, "r:gz")
                tar.extractall(os.path.join(corps_dir, corp))
                tar.close
```

**Preprocessing the data(e.g. removing punctuation).**

```python
def my_preprocessing(line):
    line = unidecode(line) #unicode the sentences
    line = line.lower() #make the sentences lowercase
    line = line.translate(str.maketrans('', '', string.punctuation)) #remove puncations
    line = line.translate(str.maketrans('', '', string.digits)) #remove numbers
    line = line.replace("\n", "") #remove new lines
    return line
    
```

**Convert sentences to ngram**

```python
def sentence_to_ngram(line, n):
    ngrams_statics = []
    for i in range(len(line) - n + 1): #making a moving window to make n-grams
        ngrams_statics.append(line[i : i + n])
    return ngrams_statics
```

**Make hyperdimensional vectors for alphabets**

```python
def hyperdimentional_alphabets(d):
    alphabets = list(string.ascii_lowercase) #take alphabets
    alphabets.append(" ") #add space(' ') to the alphabets
    hd_dic = {}
    for alphabet in alphabets: #make hyperdimentional vector for each character
        hd_dic[alphabet] = torchhd.random(1, d)
    return hd_dic

```

**Transform n-grams to hyperdimensional vectors based on the paper:<br>
"Distributed Representation of n-gram Statistics for Boosting Self-organizing Maps with Hyperdimensional Computing"**

```python
def hyperdimentional_transform(alphabets_dic, ngram):
    hdvecs_ngrams = []
    for gram in ngram: #loop over ngrams
        hdvecs_gram = []
        for char in gram: #loop over char in each grams
            idx = gram.find(char)
            hd_vec = alphabets_dic.get(char) #find the hyperdimensional presentation for the character
            for i in range(idx + 1): #loop over index of each character to permute based on its position in gram
                hd_vec = torchhd.permute(hd_vec)
            hdvecs_gram.append(hd_vec)
        hdvecs_ngrams.append(torchhd.bind(torchhd.bind(hdvecs_gram[0], hdvecs_gram[1]), hdvecs_gram[2])) #bind all the characters of a gram
    hdvec = torchhd.bundle(hdvecs_ngrams[0], hdvecs_ngrams[1]) #bundle of first two hyperdimensional vectors
    for i in range(2, len(hdvecs_ngrams) - 1): #loop over rest of hyperdimensional vectors to bundle
        hdvec = torchhd.bundle(hdvec, hdvecs_ngrams[i])
    return torch.nn.functional.normalize(hdvec).tolist()[0] #normalize between -1, 1 and convert the tensor to list and return it
        
```

**Make DataFram**

```python
def make_df(corp_count, alphabets_dic):
    corps_dir = "corps"
    corps = os.listdir(corps_dir)
    lines_list = []
    ngrams_list = []
    country_list = []
    hdvecs = []
    x = 0
    for corp in corps: #loop over countries files
        corps_files_dir = os.listdir(os.path.join(corps_dir, corp))
        for files_dir_name in corps_files_dir: #loop over directories in each country
            files_dir = os.path.join(corps_dir, corp, files_dir_name)
            if os.path.isdir(files_dir) and "wikipedia" in files_dir: #I just used the Wikipedia for my work as dataset
                 files = os.listdir(files_dir)
                 for file in files: #loop over file in the Wikipedia directory
                     if "sentence" in file: #select the sentences file
                         corp_data = []
                         file = open(os.path.join(files_dir, file), "r", encoding="utf-8") #open the file
                         l = 0
                         for line in file.readlines(): #read the file line by line
                             line = line.split('\t')[1] #each line has two part, first one is index, second one is the senteces
                             line = my_preprocessing(line) #prerocessing the sentecnes
                             ngrams = sentence_to_ngram(line, 3) #convert senteces to 3-gram
                             lines_list.append(line)
                             ngrams_list.append(ngrams)
                             country_list.append(corp)
                             hdvecs.append(hyperdimentional_transform(alphabets_dic, ngrams))
                             l += 1
                             if l >= corp_count: #read the corp_count number of the Wikipedia lines
                                 break
    df = pd.DataFrame({"line" : lines_list, "ngrams" : ngrams_list, "country" : country_list, "hdvecs" : hdvecs}) #make the datafram
    df = shuffle(df) #shuffle the data
    df = df.reset_index(drop=True) #reset the index of datafram
    dataset_df = pd.DataFrame(df.hdvecs.tolist()) #make final dataset
    dataset_df['label'] = df.country #add labels to the final dataset
    return dataset_df
```

**Train and Evaluate the model**

```python
def make_model(dataset_df):
    le = preprocessing.LabelEncoder()
    dataset_df['label'] = le.fit_transform(dataset_df.label) #convert the names of countries to integer
    X_train, X_test, y_train, y_test = train_test_split(dataset_df.iloc[:,:-1], dataset_df.iloc[:,-1], test_size=1 / 3) #split the data to train and test
    clf = RidgeClassifier(alpha=1.0) # Clarify the model
    clf.fit(X_train, y_train) #train the model
    
    predict = clf.predict(X_test) #make prediction on the test featurs
    
    conf_matrix = confusion_matrix(y_test, predict) #calculate the confusion matrix
    fscroe = f1_score(y_test, predict, average='weighted') #calcualte the F1 score
    accuracy = clf.score(X_test, y_test) #calcuate the accuracy
    fig, ax = plt.subplots(figsize=(10,10))
    sn.set(font_scale=0.5) #label size of the heatmap
    sn.heatmap(pd.DataFrame(conf_matrix), annot=True,cmap='Blues', fmt='g', annot_kws={"size": 10}) #font size of heatmap
    
    plt.show()
    print(f"F1-scroe is equal to {round(fscroe, 4) * 100}")
    print(f"Accuracy is equal to {round(accuracy,4) * 100}")
```

**Put all the function together to deploy a pipline to prepare the data, make the model, evaluate the model and report the results**

```python
def main(data_count, hdvec_dimension):
    alphabets_dic = hyperdimentional_alphabets(hdvec_dimension)
    dataset_df = make_df(data_count, alphabets_dic)
    make_model(dataset_df)
    print("\n\n")
    print("\t\t\t *** \t\t\t *** \t\t\t ***")
    print("\t\t\t *** \t\t\t *** \t\t\t ***")
    print("\n\n")
```

**Run the pipline**

```python
unextract()
for data_count, hdvec_dimension in zip([1000, 1000, 1000], [100, 1000, 10000]):
    print(f"Number of sample for each language is : {data_count}")
    print(f"The Dimension of hyperdimension vector is : {hdvec_dimension}")
    print("\n")
    main(data_count, hdvec_dimension)
```
